## Medal_Tally

```python
medal_tally=df.drop_duplicates(subset=['Team','NOC','Games','Year','Season','City','Sport','Event','Medal'])
medal_tally
```

```python
df.info()
```

```python
def visit(s):
    o=medal_tally.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False).reset_index()
    o['Total']=o['Gold']+o['Silver']+o['Bronze']
    #print(o)
    print(o[o['NOC']==s])
```

```python
print("Enter the country NOC name to calculate medals they received")
s=input()
visit(s)
```

## Medals gained when year and Team are mentioned

```python
def medals_in_year(year,Team):
    
    tempdf=medal_tally
    if(year=='overall' and Team!='overall'):
        tempdf=tempdf[tempdf['region']==Team]
        pass
    if(year!='overall' and Team=='overall'):
        tempdf=tempdf[tempdf['Year']==int(year)]
        pass
    if(year!='overall' and Team!='overall'):
        tempdf=tempdf[(tempdf['region']==Team) & (tempdf['Year']==int(year))]
        pass
    
    x=tempdf.groupby('region').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False).reset_index()
    x['Total']=x['Gold']+x['Silver']+x['Bronze']
    print(x)
    
    
     
        
```

```python
print("Enter the year")
year=(input())
print("Enter the team")
team=input()
medals_in_year(year,team)
```


## How to get Top Players of a particular sport

```python
def success(sport):
    tempdf=df.dropna(subset=['Medal'])
    tempdf=tempdf[tempdf['Sport']==sport]
    tempdf=tempdf['Name'].value_counts().reset_index().head(15).merge(df,left_on='index',right_on='Name',how='left')[['index','Name_x','Sport','region']]
    tempdf=tempdf.drop_duplicates(subset='index').rename(columns={'index':'Name','Name_x':'Medals'})
    return tempdf
    
```

```python
success('Judo')
```

## Country Wise Analysing

```python
tempdf=df.dropna(subset=['Medal'])
tempdf.drop_duplicates(subset=['Team','NOC','Games','Year','Season','City','Sport','Event','Medal'],inplace=True)
```

```python
def country(s):
    final_df=tempdf[tempdf['region']==s]
    final_df=final_df.groupby('Year')['Medal'].count().reset_index().sort_values('Year')
    plt.plot(final_df['Year'],final_df['Medal'])
    plt.xlabel("Years")
    plt.ylabel("Medals")
    plt.title("Number Of Medals a country gained in years")
    return final_df
```

```python
print("Enter the country name")
s=input()
country(s)
```

```python
x=tempdf[tempdf['region']==s]
plt.figure(figsize=(20,20))
sns.heatmap(x.pivot_table(index='Sport',columns='Year',values='Medal',aggfunc='count').fillna(0).astype('int'),annot=True)
```
